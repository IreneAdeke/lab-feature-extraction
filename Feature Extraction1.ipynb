```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv("https://raw.githubusercontent.com/ironhack-labs/lab-cleaning-categorical-data/master/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
#Categorical variables
categoricals = customer_df.select_dtypes(object)
categoricals.head()
```

```python
#Plot all the categorical variables with the proper plot. What can you see?
for var in categoricals:
    plt.figure(figsize=(16,6))
    sns.displot(data=customer_df, x=customer_df[var],  kde=True)
    plt.show()
```
